```python
# Data cleaning

""" 1. Drop these columns since they are not needed for this analysis: Gross tertiary education enrollment (%), Population, Unemployment rate,
# subscribers_for_last_30_days, Urban_population ,Latitude,Longitude """


df = df.drop(columns=["Gross tertiary education enrollment (%)","Population","Unemployment rate", "subscribers_for_last_30_days",
"Urban_population","Latitude","Longitude"])
 
```

```python
df
```

```python
# Data Cleaning

""" 2. Replace missing values in categorical columns: "category", "Abbreviation", "channel_type", "country" with "not available """

df[["category","Abbreviation","channel_type","Country"]] = df[["category","Abbreviation","channel_type","Country"]].fillna("Not Available")
```

```python
# Data Cleaning
""" 3. Replace missing values in columns: "channel_type_rank", "video_views_for_the_last_30_days" with their modes """

df[["channel_type_rank", "video_views_for_the_last_30_days", "country_rank"]] = df[["channel_type_rank", "video_views_for_the_last_30_days", "country_rank"]].apply(lambda col: col.fillna(col.mode()[0]))

```

```python
# Recheck missing values

df.isnull().sum()
```

```python
# Drop remaining missing values

df.dropna(inplace = True)
```

```python
# Next, reformat the date

df.created_date = pd.to_datetime(df.created_date)
```

```python

# Some unusual characters are still present, so these are taken care of using regular expression

df = pd.read_csv(r"C:\Users\Acer\df_clean.csv", encoding="windows-1254")  

# regex pattern to remove unusual characters
pattern = re.compile(r"[^a-zA-Z0-9\s]")

columns_to_clean = [
    "rank", "Youtuber", "subscribers", "video views", "category", "Title",
    "uploads", "Country", "Abbreviation", "channel_type",
    "video_views_rank", "country_rank", "channel_type_rank",
    "video_views_for_the_last_30_days", "lowest_monthly_earnings",
    "highest_monthly_earnings", "lowest_yearly_earnings",
    "highest_yearly_earnings", "created_year", "created_month",
    "created_date"
]

df[columns_to_clean] = df[columns_to_clean].applymap(
    lambda x: re.sub(pattern, "", str(x)) if pd.notnull(x) else x
)

# Saved the cleaned dataset to a new CSV file
df.to_csv(r"C:\Users\Acer\Desktop\Data Analysis\data\Cleaned Global YouTube Statistics.csv", encoding="windows-1254", index=False)

```
